```python
import rasterio
import matplotlib.pyplot as plt

# Define the directory and file paths for the change maps, using raw strings (r'...')
base_dir = r'D:\GLP_OSM_PC\PPT Work\Results'
change_files = {
    "Delhi": rf"{base_dir}\NDBI_change_Delhi.tif",
    "Bangalore": rf"{base_dir}\NDBI_change_banglore.tif",
    "BBR": rf"{base_dir}\NDBI_change_BBR.tif",
    "MUM": rf"{base_dir}\NDBI_change_MUM.tif"
}

# Function to plot the NDBI change map for a city
def plot_change_map(city_name, file_path):
    fig, ax = plt.subplots(figsize=(8, 6))
    
    # Load and plot the NDBI change map
    with rasterio.open(file_path) as src:
        ndbi_change = src.read(1)  # Read the first band
        extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]  # Get the spatial extent
    
    # Plot the change map using a diverging color palette
    im = ax.imshow(ndbi_change, cmap='RdYlGn_r', vmin=-0.5, vmax=0.5, extent=extent)
    ax.set_title(f"{city_name} NDBI Change (2023 - 2018)")
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Latitude")
    
    # Add a color bar to show the scale of change
    cbar = plt.colorbar(im, ax=ax, orientation='vertical', fraction=0.046, pad=0.04)
    cbar.set_label('NDBI Change')
    
    plt.tight_layout()
    plt.show()

# Plot the change maps for each city
for city, file_path in change_files.items():
    plot_change_map(city, file_path)
```

```python
# Define the directory and file paths for the change maps, using raw strings (r'...')
base_dir = r'D:\GLP_OSM_PC\PPT Work\Results'
change_files = {
    "Delhi": rf"{base_dir}\NDBI_change_Delhi.tif",
    "Bangalore": rf"{base_dir}\NDBI_change_banglore.tif",
    "BBR": rf"{base_dir}\NDBI_change_BBR.tif",
    "MUM": rf"{base_dir}\NDBI_change_MUM.tif"
}

# Create a 2x2 plot layout and export the figure as a PNG file
def plot_and_export_change_maps(change_files, output_path):
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows x 2 columns layout

    # Flatten the axes array for easier iteration
    axes = axes.flatten()
    
    # Iterate over cities and file paths to plot each change map
    for idx, (city_name, file_path) in enumerate(change_files.items()):
        with rasterio.open(file_path) as src:
            ndbi_change = src.read(1)  # Read the first band
            extent = [src.bounds.left, src.bounds.right, src.bounds.bottom, src.bounds.top]  # Get the spatial extent

        # Plot the NDBI change map using a diverging color palette
        im = axes[idx].imshow(ndbi_change, cmap='RdYlGn_r', vmin=-0.5, vmax=0.5, extent=extent)
        axes[idx].set_title(f"{city_name} NDBI Change (2023 - 2018)")
        axes[idx].set_xlabel("Longitude")
        axes[idx].set_ylabel("Latitude")

        # Add a color bar for each subplot
        cbar = plt.colorbar(im, ax=axes[idx], orientation='vertical', fraction=0.046, pad=0.04)
        cbar.set_label('NDBI Change')

    # Adjust layout for better fit
    plt.tight_layout()

    # Save the figure as a PNG file with 300 dpi
    plt.savefig(output_path, dpi=300, format='png')
    print(f"Change maps exported to {output_path} with 300 dpi.")

    # Show the plot (optional)
    plt.show()

# Define the output path for the PNG file
output_path = r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_Change_Maps_REV.png'

# Call the function to plot and export the change maps
plot_and_export_change_maps(change_files, output_path)
```

```python
# Define the directory and file paths for the NDBI maps, using raw strings (r'...')
base_dir = r'D:\GLP_OSM_PC\PPT Work\Results'
ndbi_files = {
    "Delhi": {
        "2018": rf"{base_dir}\NDBI_Delhi_2018.tif",
        "2023": rf"{base_dir}\NDBI_Delhi_2023.tif"
    },
    "Bangalore": {
        "2018": rf"{base_dir}\NDBI_banglore_2018.tif",
        "2023": rf"{base_dir}\NDBI_banglore_2023.tif"
    },
    "BBR": {
        "2018": rf"{base_dir}\NDBI_BBR_2018.tif",
        "2023": rf"{base_dir}\NDBI_BBR_2023.tif"
    },
    "MUM": {
        "2018": rf"{base_dir}\NDBI_MUM_2018.tif",
        "2023": rf"{base_dir}\NDBI_MUM_2023.tif"
    }
}

# Create a 2x2 plot layout for each city and export the figure as a PNG file
def plot_and_export_ndbi_maps(ndbi_files, output_path):
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))  # 2 rows x 2 columns layout

    # Flatten the axes array for easier iteration
    axes = axes.flatten()
    
    # Iterate over cities and file paths to plot each NDBI map
    for idx, (city_name, files) in enumerate(ndbi_files.items()):
        # Load and plot NDBI 2018
        with rasterio.open(files["2018"]) as src_2018:
            ndbi_2018 = src_2018.read(1)
            extent = [src_2018.bounds.left, src_2018.bounds.right, src_2018.bounds.bottom, src_2018.bounds.top]
        
        im1 = axes[idx].imshow(ndbi_2018, cmap='viridis', vmin=-1, vmax=1, extent=extent)
        axes[idx].set_title(f"{city_name} NDBI 2018")
        axes[idx].set_xticks([])
        axes[idx].set_yticks([])
        cbar = plt.colorbar(im1, ax=axes[idx], orientation='vertical', fraction=0.046, pad=0.04)
        cbar.set_label('NDBI Value')
        
        # Load and plot NDBI 2023
        with rasterio.open(files["2023"]) as src_2023:
            ndbi_2023 = src_2023.read(1)
        
        im2 = axes[idx].imshow(ndbi_2023, cmap='viridis', vmin=-1, vmax=1, extent=extent)
        axes[idx].set_title(f"{city_name} NDBI 2023")
        axes[idx].set_xticks([])
        axes[idx].set_yticks([])
        cbar = plt.colorbar(im2, ax=axes[idx], orientation='vertical', fraction=0.046, pad=0.04)
        cbar.set_label('NDBI Value')

    # Adjust layout for better fit
    plt.tight_layout()

    # Save the figure as a PNG file with 300 dpi
    plt.savefig(output_path, dpi=300, format='png', bbox_inches='tight')
    print(f"NDBI maps exported to {output_path} with 300 dpi.")

    # Show the plot (optional)
    plt.show()

# Define the output path for the PNG file
output_path = r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_Maps_2018_2023.png'

# Call the function to plot and export the NDBI maps
plot_and_export_ndbi_maps(ndbi_files, output_path)
```

```python
# Define the directory and file paths for Delhi's NDBI maps, using raw strings (r'...')
base_dir = r'D:\GLP_OSM_PC\PPT Work\Results'
ndbi_files = {
    "2018": rf"{base_dir}\NDBI_Delhi_2018.tif",
    "2023": rf"{base_dir}\NDBI_Delhi_2023.tif"
}

# Create a 1x2 plot layout for Delhi and export the figure as a PNG file
def plot_and_export_delhi_ndbi_maps(ndbi_files, output_path):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))  # 1 row x 2 columns layout

    # Plot NDBI 2018, keeping only positive values (built-up areas)
    with rasterio.open(ndbi_files["2018"]) as src_2018:
        ndbi_2018 = src_2018.read(1)
        extent = [src_2018.bounds.left, src_2018.bounds.right, src_2018.bounds.bottom, src_2018.bounds.top]

    # Mask out negative values to display only positive NDBI values (built-up areas)
    ndbi_2018 = np.ma.masked_less_equal(ndbi_2018, 0)
    
    im1 = axes[0].imshow(ndbi_2018, cmap='YlOrRd_r', vmin=0, vmax=1, extent=extent)
    axes[0].set_title("Delhi Built-up Areas 2018 (NDBI > 0)")
    axes[0].set_xticks([])
    axes[0].set_yticks([])
    cbar1 = plt.colorbar(im1, ax=axes[0], orientation='vertical', fraction=0.046, pad=0.04)
    cbar1.set_label('NDBI Value')

    # Plot NDBI 2023, keeping only positive values (built-up areas)
    with rasterio.open(ndbi_files["2023"]) as src_2023:
        ndbi_2023 = src_2023.read(1)
    
    # Mask out negative values to display only positive NDBI values (built-up areas)
    ndbi_2023 = np.ma.masked_less_equal(ndbi_2023, 0)
    
    im2 = axes[1].imshow(ndbi_2023, cmap='YlOrRd_r', vmin=0, vmax=1, extent=extent)
    axes[1].set_title("Delhi Built-up Areas 2023 (NDBI > 0)")
    axes[1].set_xticks([])
    axes[1].set_yticks([])
    cbar2 = plt.colorbar(im2, ax=axes[1], orientation='vertical', fraction=0.046, pad=0.04)
    cbar2.set_label('NDBI Value')

    # Adjust layout for better fit
    plt.tight_layout()

    # Save the figure as a PNG file with 300 dpi
    plt.savefig(output_path, dpi=300, format='png', bbox_inches='tight')
    print(f"NDBI maps for Delhi (built-up areas only) exported to {output_path} with 300 dpi.")

    # Show the plot (optional)
    plt.show()

# Define the output path for the PNG file
output_path = r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_Maps_Delhi_2018_2023_Builtup.png'

# Call the function to plot and export the NDBI maps for Delhi
plot_and_export_delhi_ndbi_maps(ndbi_files, output_path)
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Define the path to the downloaded Sentinel-2 image for BBR
image_path = r'D:\GLP_OSM_PC\PPT Work\Results\Sentinel\Sentinel2_BBR_2023.tif'  # Replace with your actual file path

# Define the output path for the computed NDBI map
ndbi_output_path = r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_BBR.tif'

# Compute NDBI using the formula: (SWIR - NIR) / (SWIR + NIR)
def compute_ndbi(image_path, output_path):
    # Open the Sentinel-2 image
    with rasterio.open(image_path) as src:
        # Read SWIR (B11) and NIR (B8) bands
        swir = src.read(11).astype('float32')  # B11 is typically the 11th band
        nir = src.read(8).astype('float32')  # B8 is typically the 8th band

        # Create a profile for the output GeoTIFF that matches the input file
        profile = src.profile

    # Avoid division by zero using numpy's masked arrays
    ndvi_denominator = swir + nir
    ndvi_denominator = np.where(ndvi_denominator == 0, np.nan, ndvi_denominator)  # Replace zeros with NaN

    # Calculate NDBI
    ndbi = (swir - nir) / ndvi_denominator

    # Update the profile for saving the NDBI image
    profile.update(
        dtype=rasterio.float32,
        count=1,  # We are saving a single band (the NDBI)
        compress='lzw'  # Compression to reduce file size
    )
    
    # Save the NDBI image
    with rasterio.open(output_path, 'w', **profile) as dst:
        dst.write(ndbi.astype(rasterio.float32), 1)
    
    print(f"NDBI for BBR has been saved to {output_path}")

    # Display the NDBI using matplotlib
    plt.figure(figsize=(8, 6))
    plt.imshow(ndbi, cmap='RdYlGn_r', vmin=-1, vmax=1)
    plt.colorbar(label='NDBI Value')
    plt.title("Bhubaneswar")
    plt.axis('off')
    plt.show()

# Call the function to compute NDBI
compute_ndbi(image_path, ndbi_output_path)
```

```python
import rasterio
import numpy as np

# Define the paths to the NDBI GeoTIFF files for 2018 and 2023
ndbi_paths = {
    "2018": r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_Delhi_2018.tif',  # Replace with your actual path for 2018
    "2023": r'D:\GLP_OSM_PC\PPT Work\Results\NDBI_Delhi_2023.tif'   # Replace with your actual path for 2023
}

# Function to calculate the number and percentage of built-up pixels (NDBI > 0.25)
def calculate_built_up_percentage(ndbi_path, year, threshold=0.01):
    with rasterio.open(ndbi_path) as src:
        # Read the NDBI band as a NumPy array
        ndbi = src.read(1).astype('float32')
        
        # Mask out NoData values (if any)
        ndbi = np.ma.masked_invalid(ndbi)
        
        # Calculate the total number of valid pixels (non-masked)
        total_pixels = np.ma.count(ndbi)
        
        # Calculate the number of built-up pixels where NDBI > threshold (0.25)
        built_up_pixels = np.ma.masked_less(ndbi, threshold).count()
        
        # Calculate the percentage of built-up pixels
        built_up_percentage = (built_up_pixels / total_pixels) * 100
        
        # Print the results for the specified year
        print(f"Year: {year}")
        print(f"  Total Pixels (Valid): {total_pixels}")
        print(f"  Built-up Pixels (NDBI > {threshold}): {built_up_pixels}")
        print(f"  Percentage of Built-up Pixels: {built_up_percentage:.2f}%\n")
        
        return total_pixels, built_up_pixels, built_up_percentage

# Call the function for both 2018 and 2023
for year, path in ndbi_paths.items():
    calculate_built_up_percentage(path, year)
```
